```python
# 미국 시간대에서, 한국 시간대로 반영
from datetime import datetime
import pytz  # 시간대 변환을 위한 라이브러리

# 한국 시간대 가져오기
kst = pytz.timezone('Asia/Seoul')
today = datetime.now(kst)  # 현재 시간을 한국 시간으로 변환

# 날짜와 요일 가져오기
month = today.month
day = today.day
weekday = today.strftime("%A")  # 영어 요일
weekday_korean = {
    "Monday": "월",
    "Tuesday": "화",
    "Wednesday": "수",
    "Thursday": "목",
    "Friday": "금",
    "Saturday": "토",
    "Sunday": "일"
}

# 출력
print(f"  **** {month}/{day}({weekday_korean[weekday]}) ****")
print("다음 부동산 주요 뉴스 by 지금햐.\n")

import urllib.request as ur
from bs4 import BeautifulSoup as bs
# news = 'https://news.daum.net/economy' # 분야별 뉴스 주소 입력
news = 'https://news.daum.net/estate'  # 부동산 섹션만도 잘 돌아감

soup = bs(ur.urlopen(news).read(), 'html.parser')

section = soup.find_all('div', 'box_comp box_news_headline2')  # 4개 섹션 있음

num = 1
for i in section:
    articles = i.ul.find_all('li')  # section별 기사 취합

    for i in articles:  # 기사별 타이틀과 링크 추출 출력
        print(f"[{num}] " + i.find('strong').text)  # 뉴스 타이틀

        # 해당 링크 타고 들어가서 본문/summary까지 추가하기
        soup_sel = bs(ur.urlopen(i.find('a').get('href')).read(), 'html.parser')
        for j in soup_sel.find_all('strong', 'summary_view'):
            print("  ▶", j.text)

        print("    →", i.find('a').get('href'))  # 링크 뽑기

        num += 1
        print()  # 한 줄 띄우기
```

```python
# 미국 시간대에서, 한국 시간대로 반영, 네이버 부동산 뉴스
from datetime import datetime
import pytz  # 시간대 변환을 위한 라이브러리

# 한국 시간대 가져오기
kst = pytz.timezone('Asia/Seoul')
today = datetime.now(kst)  # 현재 시간을 한국 시간으로 변환

# 날짜와 요일 가져오기
month = today.month
day = today.day
weekday = today.strftime("%A")  # 영어 요일
weekday_korean = {
    "Monday": "월",
    "Tuesday": "화",
    "Wednesday": "수",
    "Thursday": "목",
    "Friday": "금",
    "Saturday": "토",
    "Sunday": "일"
}

# 출력
print(f"  **** {month}/{day}({weekday_korean[weekday]}) ****")
print("네이버 부동산 주요 뉴스 by 지금햐.\n")

import requests
from bs4 import BeautifulSoup


header = {'User-agent' : 'Mozila/2.0'} #bot 거절 대응


url = 'https://news.naver.com/breakingnews/section/101/260' # 경제-부동산
response = requests.get(url, headers=header)

html = response.text
soup = BeautifulSoup(html, 'html.parser')

titles = soup.select('.sa_text_title._NLOG_IMPRESSION')

# HTML 내용을 파일로 저장: 추출 잘 안될 때, html 파일로 저장해서 요소확인
# with open('output.html', 'w', encoding='utf-8') as file:
#     file.write(soup.prettify())  # 들여쓰기된 HTML 저장
# print("HTML 파일로 저장되었습니다: output.html")



for title in titles:

    print(title.text.strip()) # 제목 가져오기
    print("→", title.attrs['href']) # 링크 가져오기
    print()
```
